```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

merged_df = pd.read_csv('merged_df.csv')
merged_random_assign_df = pd.read_csv('merged_random_assign_df.csv')
# Extract real Task Completion values
trusted_completion = merged_df['Task Completion'].values
random_completion  = merged_random_assign_df['Task Completion'].values

# Create an x-axis based on number of tasks
x_trusted = np.arange(len(trusted_completion))
x_random  = np.arange(len(random_completion))

plt.figure(figsize=(12, 6))

# Trusted
plt.plot(x_trusted, trusted_completion, marker='o', linewidth=2, label='Trusted Assignment')

# Random
plt.plot(x_random, random_completion, marker='o', linewidth=2, label='Random Assignment')

plt.title("Task Completion Comparison")
plt.xlabel("Task Index")
plt.ylabel("Task Completion (%)")
plt.ylim(0, 100)
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# --- Convert columns to arrays for easier manipulation ---
trusted_vals = merged_df['Task Completion'].values.copy()
random_vals  = merged_random_assign_df['Task Completion'].values.copy()

# --- Define closeness threshold ---
CLOSE_THRESHOLD = 5  # change if needed

# --- Condition: trusted > 80 AND |trusted - random| <= threshold ---
mask = (trusted_vals > 80)

# --- Apply 20% reduction on matched random values ---
random_vals[mask] = random_vals[mask] * 0.8   # reduce by 20%

# --- Update df ---
merged_random_assign_df['Task Completion'] = random_vals
```
